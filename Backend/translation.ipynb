```python
import os
from IPython.display import display, Markdown
import time
# from openai import OpenAI
import google.generativeai as genai

genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
model = genai.GenerativeModel('gemini-2.5-flash')
```

## Translation Test

```python
def translate_to_english_gemini(hindi_text):
    prompt = f"Translate the following Hindi text to English. Provide only the translation, no additional text: {hindi_text}"
    
    try:
        response = model.generate_content(prompt)
        return response.text.strip()
       
    except Exception as e:
        return f"Translation error: {e}"
```

## Translation of 2 pages (fast)

```python
def translate_first_two_pages(input_file, output_file):
    if not os.path.exists(input_file):
        print(f"Error: {input_file} not found.")
        return

    with open(input_file, "r", encoding="utf-8") as f:
        content = f.read()

    # Split content by the distinct markers you created [cite: 16, 17]
    pages = content.split("--- START PAGE ")
    translated_entries = []
    processed_count = 0

    for page in pages:
        if not page.strip() or processed_count >= 2:
            continue
            
        # Extract page number and content
        try:
            header_end = page.find(" ---")
            page_num = page[:header_end]
            page_body = page[header_end+4:].split("--- END PAGE")[0].strip()
            
            if page_body:
                print(f"Translating Page {page_num}...")
                english_text = translate_to_english_gemini(page_body)
                
                # Re-construct with markers for structured processing later 
                translated_entry = f"--- START PAGE {page_num} ---\n{english_text}\n--- END PAGE {page_num} ---"
                translated_entries.append(translated_entry)
                
                processed_count += 1
                # Sleep to prevent rate-limiting on the Free Tier API 
                time.sleep(2) 
        except Exception as e:
            print(f"Error parsing page: {e}")

    # Save specifically to the English output file
    with open(output_file, "w", encoding="utf-8") as f:
        f.write("\n\n".join(translated_entries))

# Execute for the first 2 pages
translate_first_two_pages("full_case_distinct.txt", "full_case_distinct_en.txt")
print("\nSuccess: First 2 pages translated and stored in 'full_case_distinct_en.txt'")
```

## Translating all pages

```python
# --- Logic to process and store distinct pages ---
def process_and_translate_file(input_file, output_file):
    with open(input_file, "r", encoding="utf-8") as f:
        content = f.read()

    # Splitting based on your markers to process each page sepeartely [cite: 17, 31]
    pages = content.split("--- START PAGE ")
    translated_pages = []

    for page in pages:
        if not page.strip():
            continue
            
        # Extract page number and the actual content
        header_end = page.find(" ---")
        page_num = page[:header_end]
        
        # Isolate the Hindi text between markers [cite: 12, 17]
        page_body = page[header_end+4:].split("--- END PAGE")[0].strip()
        
        if page_body:
            print(f"Translating Page {page_num}...")
            # Requirement: Programmatic API call [cite: 31, 47]
            english_text = translate_to_english_gemini(page_body)
            
            # Formatting for distinct storage [cite: 31]
            translated_entry = f"--- START PAGE {page_num} ---\n{english_text}\n--- END PAGE {page_num} ---"
            translated_pages.append(translated_entry)
            
            # Sleep to respect Free Tier rate limits (RPM/TPM) 
            time.sleep(1.5)

    # Writing the final result [cite: 51]
    with open(output_file, "w", encoding="utf-8") as f:
        f.write("\n\n".join(translated_pages))

# Execute
process_and_translate_file("full_case_distinct.txt", "full_case_distinct_en.txt")
print("Translation complete. Results stored in full_case_distinct_en.txt")
```

```python
import os
from IPython.display import display, Markdown
import time
from openai import OpenAI

openai_api = os.getenv("OPENAI_API_KEY")
client = OpenAI(
    api_key=openai_api,
)
```

```python
def translate_to_english(hindi_text, model="gpt-5-mini"):
    try:
        response = client.responses.create(
        model=model,
        instructions="Translate the following Hindi text to English. Provide only the translation, no additional text",
        input=hindi_text,
)

        print(response.output_text)

        translation = response.output_text.strip()
        return translation
        
    except Exception as e:
        return f"Translation error: {e}"
```

```python
hindi_sample = "नमस्ते, आप कैसे हैं?"

english_translation = translate_to_english(hindi_sample)
print(f"English: {english_translation}")
```
